```python
datal = pd.DataFrame([
        [58, 1.52, 7],
        [57, 1.53, 6],
        [60, 1.7, 5.8],
        [70, 1.69, 8],
        [65, 1.66, 9]
    ], columns=['Weight', 'Height', 'Score'])
datal
```

```python
datal['bias'] = [1] * datal.shape[0]
datal
```

```python
Xhat = datal[['bias','Weight', 'Height']].values
y = datal['Score'].values
```

```python
Xhat.shape, y.shape
```

$$\beta = (X^T * X)^{-1} * (X^T *y)$$

```python
np.linalg.inv(Xhat.T @ Xhat) @ (Xhat.T @ y)
```

```python
beta = np.linalg.inv(Xhat.T @ Xhat) @ (Xhat.T @ y)
beta
```

### Du doan voi weught = 56 va height = 1.4
$$\beta_0 + \beta_1 * weight + \beta_2 * height$$

```python
Xt = np.array([1, 56, 1.4])
predict = beta @ Xt.T
print('Predict', predict)
```

```python
y_predict=Xhat @ beta.T
y_predict
```

### Mean Square Error
$$mse = \frac{\sum{(y_i - y_predict)^2}}{n}$$

```python
mse = sum((y - y_predict)**2) / y.shape[0]
print('Mean Square Error', mse)
```

$$R^2 = 1 - \frac{\sum(y_1 - ypre_i)^2}{\sum(y_i - \bar{y})^2}$$

```python
r2 = 1 - sum((y - y_predict)**2) / sum((y - np.mean(y))**2)
print('R^2 score', r2)
```
